```python
#Creamos una función para renombrar la clasificación o 'rating'
def ESRB (rating):
    if rating == "E" or rating == "K-A":
        return "Everyone"
    elif rating == "T":
        return "Teens"
    elif rating == "M":
        return "Mature"
    elif rating == "E10+":
        return "Everyone 10y+"
    elif rating == "EC":
        return "Early Childhood"
    elif rating == "RP":
        return "Pending"
    elif rating == "AO":
        return "Adults Only"
    else:
        return "Pending"
```

## 5.1 Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

    - Hipotesis Nula: No existen diferencias entre las calificaciones promedio de los usuarios para las plataformas XBox One y PC.

    - Hipotesis Alternativa: Si existen diferencias entre las calificaciones promedio de los usuarios para las plataformas XBox One y PC.

```python
# Filtramos el dataframe únicamente con los videojuegos para la plataforma de XBox One y sacamos el promedio de las calificaciones de los usuarios
xone = df_games[df_games['platform'] == 'XOne']['user_score'].dropna()
xone_mean = xone.mean()
print(f'El promedio de las calificaciones de los usuarios para el XBox One es de {xone_mean:.2f} puntos.')
```

```python
# Filtramos el dataframe únicamente con los videojuegos para la plataforma de PC y sacamos el promedio de las calificaciones de los usuarios
pc = df_games[df_games['platform'] == 'PC']['user_score'].dropna()
pc_mean = pc.mean()
print(f'El promedio de las calificaciones de los usuarios para la PC es de {pc_mean:.2f} puntos.')
```

```python
## Prueba de Hipotesis

# Establecemos el nivel de significancia
alpha = 0.05

# Realizamos la prueba ttest para independientes
result = st.ttest_ind(xone, pc)

# Extraemos el valor p
print('El valor p es: ', result.pvalue)

# Comparamos el valor p con el nivel de significancia
if result.pvalue < alpha:
    print('Rechazar Hipotesis Nula')
else:
    print('No podemos rechazar la Hipotesis Nula')
```

Para realizar la prueba de hipotesis, primero obtuvimos los valores promedio tanto para las calificaciones de los usuarios del XBox One como para los de PC.
Establecimos un nivel de significancia del 5% y se extrajo el valor "p" promedio.
Calculamos una T-test para los promedios de dos muestras independientes, obteniendo un valor extremadamente pequeño e insignificante que nos conlleva a Rechazar la Hipotess Nula.

## 5.2 Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.

    - Hipotesis Nula: No existe diferencia entre las calificaciones promedio de los usuarios para los géneros de Acción y Deportes.

    - Hipotesis Alternativa: Sí existe diferencia entre las calificaciones promedio de los usuarios para los géneros de Acción y Deportes.

```python
# Filtramos el dataframe para el género de Acción, eliminamos los valores ausentes y obtenemos el promedio
action = df_games[df_games['genre'] == 'Action']['user_score'].dropna()
action_mean = action.mean()
print(f'El promedio de las calificaciones de los usuarios para el género de Acción es de {action_mean:.2f} puntos.')

# Fltramos el datafram para el género de Deportes, eliminamos los valores ausentes y obtenemos el promedio
sports = df_games[df_games['genre'] == 'Sports']['user_score'].dropna()
sports_mean = sports.mean()
print(f'El promedio de las calificaciones de los usuarios para el género de Deportes es de {sports_mean:.2f} puntos.')

# Establecemos el nivel de significancia
alpha = 0.05

# Realizamos la prueba ttest para independientes
result = st.ttest_ind(action, sports)

# Extraemos el valor p
print('El valor p es: ', result.pvalue)

# Comparamos el valor p con el nivel de significancia
if result.pvalue < alpha:
    print('Rechazar Hipotesis Nula')
else:
    print('No podemos rechazar la Hipotesis Nula')
```
